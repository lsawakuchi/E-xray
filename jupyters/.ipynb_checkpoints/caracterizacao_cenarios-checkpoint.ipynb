```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import textwrap
import matplotlib.pyplot as plt
from xray.cleaning.preprocessing import Preprocessing
from plotly import plotly
from plotly.offline import init_notebook_mode, iplot
from plotly import graph_objs as go
import matplotlib.pyplot as plt

init_notebook_mode(connected=True)
```

```python
df_cenarios = pd.read_excel("analise_escolas_final.xlsx")
```

```python
df_cenarios.shape
```

```python
df_cenarios.head()
```

```python
dfescolas = pd.read_csv("../dataset/bases_lidia/Anos finais/ideb_escola_2017_af.csv")

dfescolas.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
dfvar = dfescolas[['cod_escola_inep','num_matriculas', 'num_matriculas_ensino_integral', 'num_turmas', 'num_professores', 'in_final_semana',
                  'tp_localizacao']].copy()
```

```python
dfotimo = df_cenarios[(df_cenarios['tx_progr']>0) & (df_cenarios['tx_evolucao'] >= 0) & (df_cenarios['corr_ind']>0) 
        & (df_cenarios['ind_disp']<=5) & (df_cenarios['ideb_ai']>=6) & (df_cenarios['ideb_af']>=6)]
```

```python
dfotimo = dfotimo.merge(dfvar, left_on='escola', right_on='cod_escola_inep', how='left')
```

```python
dfotimo['cenario'] = 'otimo'
```

```python
df_evoluindo = df_cenarios[(df_cenarios['tx_progr']>0) & (df_cenarios['tx_evolucao'] >= 0) & (df_cenarios['corr_ind']>0)
                           & (df_cenarios['ind_disp']<=5) & (df_cenarios['ideb_ai']<6) & (df_cenarios['ideb_af']<6)]
```

```python
df_evoluindo.shape
```

```python
df_evoluindo = df_evoluindo.merge(dfvar, left_on='escola', right_on='cod_escola_inep', how='left')
```

```python
df_evoluindo['cenario'] = 'evoluindo'
```

```python
df_queda = df_cenarios[(df_cenarios['ind_disp']< 5) & (df_cenarios['tx_progr'] < 0)]
```

```python
df_queda.shape
```

```python
df_queda = df_queda.merge(dfvar, left_on='escola', right_on='cod_escola_inep', how='left')
```

```python
df_queda['cenario'] = 'queda'
```

```python
df_atencao = df_cenarios[df_cenarios['ind_disp']>5]
```

```python
df_atencao['cenario']='atencao'
```

```python
df_atencao.shape
```

```python
df_atencao = df_atencao.merge(dfvar, left_on='escola', right_on='cod_escola_inep', how='left')
```

```python
df = pd.concat([dfotimo, df_evoluindo, df_queda, df_atencao])
```

```python
df
```

```python
df.columns
```

```python
df = df[['escola', 'cenario', 'in_final_semana', 'num_matriculas', 'num_matriculas_ensino_integral', 'num_professores', 
    'num_turmas', 'tp_localizacao', 'tp_rede', 'uf']].copy()
```

```python
df_atv = df.groupby(["cenario", "in_final_semana"]).count().reset_index()[['cenario', 'in_final_semana', 'escola']]
```

```python
df_atv
```

```python
df_total = df_atv.groupby('cenario').sum().reset_index()[['cenario', 'escola']]
```

```python
df_total.columns = ['cenario', 'total']
```

```python
df_atv = df_atv.merge(df_total, left_on='cenario', right_on='cenario', how='left')
```

```python
df_atv['prop'] = df_atv['escola']/df_atv['total']
```

```python
df_atv
```

```python
df_atv[['cenario', 'in_final_semana','prop']]
```

```python
color_opt = ['rgb(21,201,190)', 'rgb(132,205,137)']
```

```python
data = []
df1 = df_atv[df_atv['in_final_semana']==1]
df1 = df1.sort_values('prop', ascending=False)
trace1 = go.Bar(
    x = df1['cenario'].unique().tolist(),
    y = df1['prop']*100,
    name = 'oferece',
    marker = dict(color=color_opt[0], line=dict(color='rgb(0, 0, 0)', width=1))
)
df2 = df_atv[df_atv['in_final_semana']==0]
df2 = df2.sort_values('prop', ascending=False)
trace2 = go.Bar(
    x = df2['cenario'].unique().tolist(),
    y = df2['prop']*100,
    name = 'no oferece',
    marker = dict(color=color_opt[1], line=dict(color='rgb(0, 0, 0)', width=1))
)

layout = go.Layout(title = "Varivel : Oferece atividades aos finais de semana", barmode='stack', bargap=0.5, 
                   yaxis = dict(tickmode='array', 
                                tickvals = [0, 20, 40, 60, 80, 100],
                                ticktext=['0%', '20%', '40%', '60%', '80%', '100%']),
                  annotations = [dict(x = 'atencao', y = 40, xref='x', yref='y', text='79.0%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                 dict(x = 'atencao', y = 90, xref='x', yref='y', text='21.0%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                dict(x = 'queda', y = 40, xref='x', yref='y', text='77.5%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                dict(x = 'queda', y = 90, xref='x', yref='y', text='22.5%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                dict(x = 'evoluindo', y = 40, xref='x', yref='y', text='67.5%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                dict(x = 'evoluindo', y = 90, xref='x', yref='y', text='32.5%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                dict(x = 'otimo', y = 40, xref='x', yref='y', text='59%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                dict(x = 'otimo', y = 90, xref='x', yref='y', text='41%', showarrow=False, font=dict(color='rgb(255, 255, 255)')),
                                ])
fig = go.Figure(data=[trace2, trace1], layout=layout)
iplot(fig)
```

```python
dfvar['mat_prof'] = dfvar['num_matriculas']/dfvar['num_professores']
dfvar['prof_turma'] = dfvar['num_professores']/dfvar['num_turmas']
```

```python
data = []
color_menu = ['rgb(12,79,202)', 'rgb(5,171,161)', 'rgb(92,210,139)',  'rgba(7,249,186, 0.8)']

i = 0
for cen in df['cenario'].unique().tolist():
    dt = df[df['cenario']==cen]
    trace = go.Box(
        y = dt['mat_prof'],
        name = cen,
        boxmean = 'sd',
        marker = dict(color=color_menu[i]),
        #boxpoints='suspectedoutliers',
        #orientation='h',
        #boxpoints='all'
    
    )
    data.append(trace)
    i+=1
trace2 = go.Scatter(
    x = ['otimo', 'evoluindo', 'queda', 'atencao'],
    y = [18.92]*4,
    line = dict(color = 'rgb(227,21,21)', dash='dash'),
    name = "mdia geral"
)
data.append(trace2)
layout = go.Layout(title = "Distribuio da relao matricula/professor por cenrio", yaxis=dict(title="matriculas/professores"))
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
